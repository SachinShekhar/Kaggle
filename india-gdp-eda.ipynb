## Univariate Analysis

```python
# Helper functions for Numerical Univariate Analysis
def num_uni_box_analysis(df, var, friendly_name):
  ax = sns.boxplot(data=df, x=var)
  ax.set_xticks(list(df[var].quantile([0,0.25,0.5,0.75,0.95,1])))
  ax.set_xlabel(friendly_name)
  ax.set_title(f'{friendly_name} Distribution Univariate Analysis', fontsize=20)
  ax.figure.set_size_inches(24,8)

def num_uni_hist_analysis(df, var, friendly_name):
  ax = sns.histplot(data=df ,x=var, stat='percent')
  ax.set_xlabel(friendly_name)
  ax.set_title(f'{friendly_name} Distribution Histogram', fontsize=20)
  ax.figure.set_size_inches(24,8)
```

### Feature: Year

```python
num_uni_box_analysis(df, 'Year', 'Year')
plt.show()
```

The 1991 being the median makes sense as this is a 60 years data.

```python
num_uni_hist_analysis(df, 'Year', 'Year')
plt.show()
```

This uniform distribution makes sense as data was collected every year.

### Feature: GDP_In_Billion_USD

```python
num_uni_box_analysis(df, 'GDP_In_Billion_USD', 'GDP (in Billion USD)')
plt.show()
```

```python
num_uni_hist_analysis(df, 'GDP_In_Billion_USD', 'GDP (in Billion USD)')
plt.show()
```

Within the last 60 years, India's GDP has been very low most of the time.

### Feature: Per_Capita_in_USD

```python
num_uni_box_analysis(df, 'Per_Capita_in_USD', 'Per Capita Income (in USD)')
plt.show()
```

```python
num_uni_hist_analysis(df, 'Per_Capita_in_USD', 'Per Capita Income (in USD)')
plt.show()
```

Within the last 60 years, India's per capita income has been very low most of the time.

### Feature: Percentage_Growth

```python
num_uni_box_analysis(df, 'Percentage_Growth ', 'Percentage Growth')
plt.show()
```

```python
num_uni_hist_analysis(df, 'Percentage_Growth ', 'Percentage Growth')
plt.show()
```

Within the last 60 years, India's GDP growth rate has been high most of the time.
